```python
import os
from pathlib import Path
import pandas as pd


__file__ = os.path.realpath("__file__")
root_path = Path(__file__).parent
data_path = root_path / "data"

for path in data_path.glob("**/*"):
    if path.is_file():
        print(f"file: {path}")
    elif path.is_dir():
        print(f"dir : {path}")

billboardhot100_data_path = data_path / "kaggle" / "billboardHot100_1999-2019.csv"
df = pd.read_csv(billboardhot100_data_path)
print(df.info())
df.head()
```

```python
years = list(range(1999, 2020))
billboard_genre_by_year = {}
for year in years:
    billboard_genre_by_year[year] = []

for row in df.itertuples():
    try:
        year = row.Week[:4]
    except:
        continue
    billboard_genre_by_year[int(year)].extend(row.Genre.upper().split(","))

for key in billboard_genre_by_year.keys():
    billboard_genre_by_year[key] = list(set(billboard_genre_by_year[key]))

print("billboard")
for key in billboard_genre_by_year.keys():
    print(f"{key}{len(billboard_genre_by_year[key]):>8}")
```

```python
spotify_genre_by_year = {}
for year in years:
    f = open(data_path / "everynoise" / "a_retromatic_history_of_music" / (str(year) + ".txt"), "r")
    spotify_genre_by_year[year] = [line[:-1].upper() for line in f.readlines()]
    f.close()

print("spotify")
for key in spotify_genre_by_year.keys():
    print(f"{key}{len(spotify_genre_by_year[key]):>8}")
```

```python
print("billboard (&) spotify")
for year in years:
    print(f"{year}{len(set(billboard_genre_by_year[year]) & set(spotify_genre_by_year[year])):>8}")
```
